```python
# Importing all libraries required in this notebook

import pandas as pd
import numpy as np  
import matplotlib.pyplot as plt  
%matplotlib inline

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression as regr
import seaborn as sns
```

```python
# Reading the Data

data = pd.read_csv('http://bit.ly/w-data')
print("Data imported!")

data.head()
```

Plotting the Graph of the given data:

```python
# Plotting the distribution of scores

data.plot(x='Hours', y='Scores', style='o')  
plt.title('Hours vs Percentage', size=20)  
plt.xlabel('No. of hours Studied')  
plt.ylabel('Percentage Scores') 
plt.grid()
plt.show()
```

Plotting Linear Regression Plot:

```python
plt.title('Linear Reression Plot', size=20)  
sns.regplot(x= data['Hours'], y= data['Scores'])
plt.ylabel('Marks Percentage')
plt.xlabel('Hours Studied')
plt.grid()
```

### Training the Model

```python
# Defining X and y from the Data
X = data.iloc[:, :-1].values  
y = data.iloc[:, 1].values

# Spliting the Data in two training and test sets
train_X, test_X, train_y, test_y = train_test_split(X, y, random_state = 0)
```

```python

regr = LinearRegression()
regr.fit(train_X, train_y)
print("Our Model is Trained!")  
```

#### Prediction Algorithm:-

```python
#Predicting the marks - In hours

pred_y = regr.predict(test_X)
prediction = pd.DataFrame({'Hours': [i[0] for i in test_X], 'Predicted Scores': [j for j in pred_y]})
prediction[:6]
```

```python
# Comparing Actual vs Predicted
df = pd.DataFrame({'Actual': test_y, 'Predicted': pred_y})  
df 
```

### Evaluating the Model
Calculating the accuracy of the regression model:

```python
# Finding the mean error
from sklearn.metrics import mean_absolute_error
print("The Mean absolute error of the Model is : ",mean_absolute_error(test_y,pred_y))
```

### Predicting the scores of a student if studied time is 9.25hrs:

```python
hrs = [9.25]
result = regr.predict([hrs])
print(f"No of Hours studied: {hrs}")
print(f"Predicted Score = {result}")
```

### *According to our trained model, predicted score is 93.89 after studing for 9.25 hrs.*
